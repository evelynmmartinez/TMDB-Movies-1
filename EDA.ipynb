```python
#imports
import pandas as pd
```

```python
df_00 = pd.read_csv('Data/final_tmdb_data_2000.csv.gz')
df_00.head()
```

```python
df_01 = pd.read_csv('Data/final_tmdb_data_2001.csv.gz')
df_01.head()
```

```python
dfs = [df_00,df_01]
df1 = pd.concat(dfs)
```

1. How many movies had at least some valid financial information (values > 0 for budget OR revenue)?

```python
budget_over_zero = df1[(df1['budget'] > 0) | (df1['revenue']>0)]
movies_required_budget =len(budget_over_zero)
print("Amount of movies that had at least some valid financial information (values > 0 for budget OR revenue",movies_required_budget)
```

2. How many movies are there in each of the certification categories (G/PG/PG-13/R)?


```python

certifications = df1['certification'].value_counts()
print("Movies in each certification category:" , certifications)
print(certifications)
```

3. What is the average revenue per certification category?

```python
avg_rev_by_certification = df1.groupby('certification')['revenue'].mean()
print("Avg revenue per certification category:")
print(avg_rev_by_certification)
```

4. What is the average budget per certification category?

```python
avg_budget_by_certification = df1.groupby('certification')['budget'].mean()
print("Avg budget per certification category:")
print(avg_budget_by_certification)
```

```python
df1.to_csv("Data/tmdb_results_combined.csv.gz", compression='gzip', index=False)
df1.head()
```
